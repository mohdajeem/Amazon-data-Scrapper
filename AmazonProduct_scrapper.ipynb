```python
from selenium import webdriver
from bs4 import BeautifulSoup
from tqdm import tqdm
import pandas as pd
import numpy as np
```

```python
data = [] 
for sp in tqdm(soup.find_all('div', class_ = 'a-section a-spacing-base')):
    try:
        prod_photo = sp.find('img').get('src')
    except:
        prod_photo = np.nan
    try:
        prod_link = 'https://www.amazon.in'+sp.find('a').get('href')
    except:
        prod_link = np.nan
    try:
        prod_name = sp.find('span', class_ = 'a-size-base-plus a-color-base a-text-normal').text
    except:
        prod_name = np.nan
    try:
        prod_price = sp.find('span', class_ = 'a-offscreen').text[1:]
    except:
        prod_price = np.nan
    if(sp.find('span', class_ = 'a-size-base a-color-secondary').text.split(' ')[0] == 'M.R.P:'):
        sale_last_month = '0'
    else:
        sale_last_month = sp.find('span', class_ = 'a-size-base a-color-secondary').text.split(' ')[0]

        
    if(sp.find('span', class_ = 'a-declarative').text[:9] == 'Sponsored'):
        sponsored = sp.find('span', class_ = 'a-declarative').text[:9]
    else:
        sponsored = 'Not-Sponsored'
    # prod_photo = sp.find('img').get('src')
    # prod_link = 'https://www.amazon.in'+sp.find('a').get('href')
    # prod_name = sp.find('span', class_ = 'a-size-base-plus a-color-base a-text-normal').text
    # prod_price = sp.find('span', class_ = 'a-offscreen').text[1:]
    # print(sp.find('span', class_ = 'a-size-base-plus a-color-base a-text-normal').text)
    for sp1 in sp.find_all('div', class_ = 'a-row a-size-small'):
        try:
            rating = sp1.find_all('span')[0].text[:3]
        except:
            rating = np.nan
        try:
            no_of_ratings = sp1.find_all('a')[1].text
        except:
            no_of_ratings = np.nan
        
        break

    data.append([prod_photo, prod_link, prod_name, prod_price, rating, no_of_ratings,sponsored,sale_last_month])
    
```

```python
browser = webdriver.Chrome()
data = []
for i in tqdm(range(1,21)):
    link = 'https://www.amazon.in/s?k=oneplus+buds&page='+str(i)+'&crid=2IZGJF2AM2DKC&qid=1725338612&sprefix=oneplus+buds%2Caps%2C513&ref=sr_pg_2'
    browser.get(link)
    soup = BeautifulSoup(browser.page_source, 'html.parser')
    for sp in soup.find_all('div', class_ = 'a-section a-spacing-base'):
        try:
            prod_photo = sp.find('img').get('src')
        except:
            prod_photo = np.nan
        try:
            prod_link = 'https://www.amazon.in'+sp.find('a').get('href')
        except:
            prod_link = np.nan
        try:
            prod_name = sp.find('span', class_ = 'a-size-base-plus a-color-base a-text-normal').text
        except:
            prod_name = np.nan
        try:
            prod_price = sp.find('span', class_ = 'a-offscreen').text[1:]
        except:
            prod_price = np.nan
        try:
            if(sp.find('span', class_ = 'a-size-base a-color-secondary').text.split(' ')[0] == 'M.R.P:'):
                sale_last_month = '0'
            else:
                sale_last_month = sp.find('span', class_ = 'a-size-base a-color-secondary').text.split(' ')[0]
        except:
            sale_last_month = np.nan
        
        try:
            if(sp.find('span', class_ = 'a-declarative').text[:9] == 'Sponsored'):
                sponsored = sp.find('span', class_ = 'a-declarative').text[:9]
            else:
                sponsored = 'Not-Sponsored'
        except:
            sponsored = np.nan
        # prod_photo = sp.find('img').get('src')
        # prod_link = 'https://www.amazon.in'+sp.find('a').get('href')
        # prod_name = sp.find('span', class_ = 'a-size-base-plus a-color-base a-text-normal').text
        # prod_price = sp.find('span', class_ = 'a-offscreen').text[1:]
        # print(sp.find('span', class_ = 'a-size-base-plus a-color-base a-text-normal').text)
        for sp1 in sp.find_all('div', class_ = 'a-row a-size-small'):
            try:
                rating = sp1.find_all('span')[0].text[:3]
            except:
                rating = np.nan
            try:
                no_of_ratings = sp1.find_all('a')[1].text
            except:
                no_of_ratings = np.nan
            
            break
    
        data.append([prod_photo, prod_link, prod_name, prod_price, rating, no_of_ratings,sponsored,sale_last_month])
    
```

```python
df = pd.DataFrame(data, columns = ['prod_photo', 'prod_link', 'prod_name', 'prod_price', 'rating', 'no_of_ratings','sponsored','sale_last_month'])
```
